```python
import pandas as pd
from model_lab.quick_score_card import QuickScoreCard
from numpy import infty

data = pd.read_excel('test_data.xlsx')
data['PassengerId'] = list(range(1, len(data) + 1))
data = data.set_index('PassengerId')

dataset_config = {
    'label_col_name': 'Survived',
    'dataset_col_name': 'dataset_name',
    'train_dataset_name': 'train',
    'test_dataset_names': ['test'],
    'oot_dataset_name': 'oot'
}

auto_binning_config = {
    'bin_max_all': 4,
    'min_sample_n': 100
}

user_config = {
    'Age': {
        'mode': 'manual',
        'keep_values': [],
        'bins': [-infty, 19, 42, infty]
    }
}


trainer = QuickScoreCard(data, dataset_config, auto_binning_config)
trainer.set_features_performance_threshold(0.01, 0.1, 0.1)
trainer.set_corr_threshold_for_model(0.7)
trainer.run()
trainer.display_summary()
trainer.to_excel('test_1')
```
